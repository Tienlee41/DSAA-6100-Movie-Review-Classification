```python
# import libraries
import pandas as pd

import nltk

from nltk.sentiment.vader import SentimentIntensityAnalyzer

from nltk.corpus import stopwords

from nltk.tokenize import word_tokenize

from nltk.stem import PorterStemmer, WordNetLemmatizer


# Load the amazon review dataset

df_predict = pd.read_csv('../data/test_data.csv')

df_predict
```

```python
# create preprocess_text function
def preprocess_text(text):
    # Tokenize the text
    tokens = word_tokenize(text.lower())

    # Remove stop words
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]

    # Steamming the tokens
    stemmer = 

    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]


    # Join the tokens back into a string

    processed_text = ' '.join(lemmatized_tokens)
    return processed_text

# apply the function df
df_predict['text'] = df_predict['text'].apply(preprocess_text)
df_predict
```

```python
# initialize NLTK sentiment analyzer

analyzer = SentimentIntensityAnalyzer()


# create get_sentiment function

def get_sentiment(text):

    scores = analyzer.polarity_scores(text)

    # # sentiment = 1 if scores['pos'] > 0.2 else 0

    # return sentiment
    return scores['pos']
# apply get_sentiment function

df_predict['Category'] = df_predict['text'].apply(get_sentiment)

df_predict
```

```python
point = df_predict['Category'].sum()/10000
```

```python
point
```

```python
def get_sentiment_point(text):

    scores = analyzer.polarity_scores(text)

    sentiment = 1 if scores['pos'] >= point else 0

    return sentiment
    # return scores['pos']
# apply get_sentiment function

df_predict['Category'] = df_predict['text'].apply(get_sentiment_point)

df_predict
```
